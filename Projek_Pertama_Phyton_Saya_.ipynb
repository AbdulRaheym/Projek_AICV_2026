Games Teka Nombor

```python
import random

# Komputer pilih nombor rawak antara 1 hingga 20
nombor_rahsia = random.randint(1, 20)
tekaan = 0

print("Jom main! Saya dah pilih satu nombor antara 1 hingga 20.")

# Gelung akan berjalan selagi tekaan salah
while tekaan != nombor_rahsia:
    tekaan = int(input("Masukkan tekaan anda: "))

    if tekaan < nombor_rahsia:
        print("Terlalu rendah! Cuba lagi.")
    elif tekaan > nombor_rahsia:
        print("Terlalu tinggi! Cuba lagi.")
    else:
        print("Tahniah! Tekaan anda betul! ")
```

# **Kalkulator Umur Kucing**

```python
print("--- KALKULATOR UMUR KUCING ---")

# Ambil input nama dan umur kucing
nama_kucing = input("Masukkan nama kucing anda: ")
umur_manusia = input("Masukkan umur kucing tersebut (dalam tahun manusia): ")

# Tukar input teks (str) kepada nombor (int) untuk dikira
umur_int = int(umur_manusia)

# Formula mudah: 1 tahun manusia bersamaan 7 tahun kucing
umur_kucing = umur_int * 7

print(f"Wah, umur {nama_kucing} dalam tahun kucing adalah {umur_kucing} tahun! ")
```
