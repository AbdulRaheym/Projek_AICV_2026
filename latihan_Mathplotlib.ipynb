```python
import matplotlib.pyplot as plt

# data
nama_buah = ['epal','Oren','Lemon']
jualan = [10,15,5]

# carta bar
plt.bar(nama_buah, jualan,color=['red','orange','yellow'])

# tajuk dan label
plt.title("Jualan Buah Hari Ini")
plt.xlabel("Nama Buah")
plt.ylabel("Bilangan Jualan")
plt.yticks(range(0,16,1))

# paparkan graf
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

x = [0, 1, 2]
y = [0, 0, 0]
z = [0, 0, 0]

dx = [0.5, 0.5, 0.5]
dy = [0.5, 0.5, 0.5]
dz = [10, 15, 5]

warna = ['red', 'orange', 'yellow']

ax.bar3d(x, y, z, dx, dy, dz, color=warna)

ax.set_xticks([0.25, 1.25, 2.25])
ax.set_xticklabels(['Epal', 'Oren', 'Lemon'])

ax.set_xlabel('Buah')
ax.set_ylabel('')
ax.set_zlabel('Jualan')
ax.set_title('Jualan Buah (3D)')

plt.show()
```
